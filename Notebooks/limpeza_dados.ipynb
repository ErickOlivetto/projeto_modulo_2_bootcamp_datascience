```python
import pandas as pd
```

```python
def arruma_df(banco_dados:pd.DataFrame, regioes:dict):
  banco_dados = banco_dados.drop('Total', axis = 1)
  banco_dados = banco_dados.melt('Unidade da Federação')
  banco_dados = banco_dados.rename(columns = {'Unidade da Federação':'uf', 'variable' : 'ano', 'value' : 'valor'})
  banco_dados['regiao'] = banco_dados['uf'].str[0].map(regioes)
  banco_dados = banco_dados.set_index('uf')
  banco_dados['ano'] = banco_dados['ano'].astype('int32')
  banco_dados.index = banco_dados.index.str[3:]
  banco_dados = banco_dados.dropna()
  return banco_dados
```

```python
url_cobertura_vacinal = 'https://github.com/ErickOlivetto/projeto_modulo_2_bootcamp_datascience/raw/main/Dados/cobertura_vacinal_uf.csv'
url_calculo_cv = 'https://github.com/ErickOlivetto/projeto_modulo_2_bootcamp_datascience/raw/main/Dados/calculo_cv_uf.csv'
```

```python
dados_cobertura_vacinal = pd.read_csv(url_cobertura_vacinal, sep = ';', skiprows = 3, engine = 'python', skipfooter = 20, decimal = ',', na_values = '...')
dados_cobertura_vacinal
```

```python
dados_calculo_cv = pd.read_csv(url_calculo_cv, sep = ';', skiprows = 3, engine = 'python', skipfooter = 20, encoding = 'ISO-8859-1', na_values = '-')
dados_calculo_cv
```

```python
regioes = {'1':'Norte', '2':'Nordeste', '3':'Sudeste', '4':'Sul', '5':'Centro-Oeste'}
dados_cobertura_vacinal = arruma_df(dados_cobertura_vacinal, regioes)
dados_calculo_cv = arruma_df(dados_calculo_cv, regioes)
```

```python
dados_cobertura_vacinal.to_csv('./dados_cobertura_vacinal.csv')
dados_calculo_cv.to_csv('./dados_calculo_cv.csv')
```
